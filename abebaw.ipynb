1) Bubble Sort is a basic sorting algorithm that:

Repeatedly compares adjacent elements and swaps them if they’re in the wrong order.

After each full pass through the list, the largest unsorted item “bubbles up” to its correct position.

It is easy to understand and implement, but it’s not efficient for large datasets due to its O(n²) time complexity.

Sorting by Student ID with Bubble Sort works well for small lists like class rosters or group assignments.

It’s stable—students with the same ID (in theory) won’t have their order mixed up.

But for large numbers of students (e.g., in a university database), better algorithms like Merge Sort or Timsort are recommended.



```python
def bubble_sort_by_id(student_list):
    n = len(student_list)
    for i in range(n):
        for j in range(0, n - i - 1):
            if student_list[j]['id'] > student_list[j + 1]['id']:
                # Swap if IDs are out of order
                student_list[j], student_list[j + 1] = student_list[j + 1], student_list[j]
    return student_list

# Example usage:
sorted_students = bubble_sort_by_id(students.copy())
for s in sorted_students:
    print(s)
```

2) Insertion Sort works like how we sort playing cards:

It builds the sorted list one item at a time.

Each new student (card) is compared to the ones before it and placed in the correct position.

✅ Why Use Insertion Sort for Names?

It’s efficient for small or nearly sorted datasets.

Great for sorting student names alphabetically.

It's a stable sort, meaning students with the same name will maintain their original order.

Not suitable for large datasets due to its O(n²) worst-case time complexity.



```python
def insertion_sort_by_name(student_list):
    for i in range(1, len(student_list)):
        key = student_list[i]  # Current student
        j = i - 1
        # Compare names in previous part of the list
        while j >= 0 and student_list[j]['name'] > key['name']:
            student_list[j + 1] = student_list[j]
            j -= 1
        student_list[j + 1] = key  # Place key in its correct position
    return student_list

# Example usage:
sorted_by_name = insertion_sort_by_name(students.copy())
for s in sorted_by_name:
    print(s)
```

3) Selection Sort works by:

Finding the maximum (or minimum) element in the unsorted part of the list,

Placing it in its correct position in the sorted part.

It continues this process until the entire list is sorted.

✅ Why Use Selection Sort for CGPA?

It’s simple and easy to implement.

Works fine for small datasets like classroom records.

Perfect for sorting students by CGPA from highest to lowest.

But: it’s not stable and inefficient for large datasets, with a time complexity of O(n²).



```python
def selection_sort_by_cgpa(student_list):
    n = len(student_list)
    for i in range(n):
        # Assume the current index has the highest CGPA
        max_index = i
        for j in range(i + 1, n):
            if student_list[j]['cgpa'] > student_list[max_index]['cgpa']:
                max_index = j
        # Swap the found max with the first unsorted element
        student_list[i], student_list[max_index] = student_list[max_index], student_list[i]
    return student_list
```


🔹 Bubble Sort (Sorting by ID)

How it works:   It repeatedly compares each pair of adjacent elements and swaps them if they’re in the wrong order. After each full pass, the largest unsorted element “bubbles” to the end.

Best case performance: Fast when the list is already sorted (linear time, O(n)).

Worst case performance: Very slow if the list is in reverse order (O(n²)).

Stable: Yes — maintains the order of equal elements.

When to use: Only for small datasets or educational purposes. Not efficient in practice.

🔹 Insertion Sort (Sorting by Name)

How it works: Builds the sorted list one item at a time by placing each new item into the correct position relative to already sorted items.

Best case performance: Excellent if the list is already or nearly sorted (O(n)).

Worst case performance: Slows down with large, unsorted data (O(n²)).

Stable: Yes — keeps the order of items with the same key.

When to use: Great for small lists or nearly sorted data. Simple and efficient in such cases.

🔹 Selection Sort (Sorting by CGPA Descending)

How it works: Repeatedly finds the highest (or lowest) item from the unsorted portion and moves it to its correct position.

Best and worst case performance: Always O(n²), no matter how sorted the list already is.

Stable: No — may change the order of equal items.

When to use: Useful for small lists when you want to minimize the number of swaps. But overall not recommended for larger data.

✅ Final Thoughts
Insertion sort is the best among the three if your list is nearly sorted or relatively short.

Bubble sort is intuitive but inefficient, good only for learning.

Selection sort is simple and uses minimal swaps, but it's not stable and not efficient for large data.

Advantage and Disadvantage of each algorithim

Used to sort by: Student ID

Advantages:

Very simple and easy to understand.

Can be implemented quickly for basic tasks.

Works well when the data is almost sorted.

Stable sort: maintains the relative order of equal keys.

Extremely inefficient for large datasets due to O(n²) time.

Performs many unnecessary comparisons and swaps.

Not ideal when performance matters.

In dictionary data: Bubble Sort can be used to sort by any key (like 'id') but becomes inefficient as data size increases.

Used to sort by: Name (alphabetical order)

Advantages:

Efficient for small datasets or nearly sorted lists.

Stable sort: keeps original order of equal values.

Easy to modify for sorting by different dictionary keys.

Good when new items are inserted frequently (like real-time sorting).

Disadvantages:

Still has O(n²) time complexity in the worst case.

Not suitable for large, unsorted datasets.

Becomes slower as data grows.

In dictionary data: Insertion Sort is useful for sorting by keys like 'name' if the list is small or nearly sorted.

🔹 Selection Sort

Used to sort by: CGPA (descending)

Advantages:

Simple to implement and understand.

Performs fewer swaps than Bubble Sort.

No need to check if the list is already sorted.

Can sort by any key like 'cgpa', 'age', etc.

Disadvantages:

Not stable: equal elements might lose their original order.

Inefficient for large datasets (O(n²) time).

Still does a lot of comparisons.

In dictionary data: Selection Sort is decent for sorting by numeric values like 'cgpa', but it’s not stable and not recommended for large data.

🧠 Summary
